```python

import ast

# Lista para almacenar todas las palabras extraídas
todas_las_palabras = []

# Función para extraer palabras entre comillas simples
def extract_words(row):
    try:
        # Convierte la cadena de texto en una lista de Python usando ast.literal_eval
        lista = ast.literal_eval(row)
        
        # Filtra las palabras que están entre comillas simples
        palabras_extraidas = [palabra.strip("'") for palabra in lista]
        
        # Agrega las palabras a la lista global
        todas_las_palabras.extend(palabras_extraidas)
    except ValueError:
        pass

# Aplica la función a la columna 'tu_columna' para recorrer todo el DataFrame
todo['genres'].apply(extract_words)

# Muestra la lista resultante de todas las palabras extraídas
print(todas_las_palabras)


```

```python
# Convierte la lista de todas_las_palabras en un conjunto para eliminar duplicados
palabras_unicas_set = set(todas_las_palabras)

# Convierte el conjunto nuevamente en una lista
palabras_unicas = list(palabras_unicas_set)

# Muestra la lista resultante de palabras únicas
print(palabras_unicas)
```

```python
from collections import Counter
from itertools import chain

def top_generos(columna, n=10):
    # Dividir las palabras en cada fila y contar la frecuencia de cada palabra
    contar = Counter(chain.from_iterable(columna.str.split(', ')))

    # Obtener las n palabras más comunes y quitar comillas dobles y corchetes
    palabras_comunes = [palabra.replace("'", "").strip("[]") for palabra, _ in contar.most_common(n)]

    return palabras_comunes

# Ejemplo de uso:
# Supongamos que df es tu DataFrame y 'tu_columna' es la columna que contiene las listas de palabras.
# Reemplaza df['tu_columna'] con el nombre real de tu columna.
resultados = top_generos(steam_games['genres'])
print(resultados)
```
