```python
import pandas as pd
```

```python
steam_games = pd.read_csv('../data/limpio/steam_games.csv.gz',compression='gzip')
users_items = pd.read_csv('../data/limpio/users_items.csv.gz',compression='gzip')
user_reviews = pd.read_csv('../data/limpio/user_reviews.csv.gz',compression='gzip')
```

```python
steam_games=steam_games[['app_name','genres','id','year','developer']]
user_reviews=user_reviews['item_id'].drop_duplicates()
users_items=users_items['item_id'].drop_duplicates()
```

```python
todo=steam_games.merge(user_reviews, left_on='id',right_on='item_id',how='inner')
```

```python
todo=todo.merge(users_items, on='item_id', how='inner')
```

```python
todo = todo.drop('id', axis=1)
```

```python
todo['genres'] = todo['genres'].apply(lambda x: x.strip("[]'").split(', ') if '[' in x else [x.strip("[]'")])
```

```python
todo
```

```python
def clean_genre_list(genre_list):
    return [palabra.strip(" '") for palabra in genre_list]
```

```python
todo['genres'] = todo['genres'].apply(clean_genre_list)
```

```python
todo
```

```python
from collections import Counter
from itertools import chain
contar = Counter(chain.from_iterable(todo['genres']))
```

```python
contar = contar.most_common(2)
```

```python
contar
```

```python
def drop_not_important(genres_list):
    return ', '.join([palabra for palabra in genres_list if palabra in [cont for cont, _ in contar]])
```

```python
todo['genres'] = todo['genres'].apply(drop_not_important)
```
